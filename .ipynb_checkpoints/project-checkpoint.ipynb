```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x= "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from above chart we can see that:
#no. of females in the data  is more than no. of males.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Relation b\w parents's education and students scores.")
plt.show()
```

```python
#from mthe above chart we conclude that education of parents have a good impact on students scores.
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot = True)
plt.title("Relation b\w parents's marital status and studesnts scores")
plt.show()
```

```python
#from the above chart we conclude that parents marital status has neglegibal impact on students score. 
```

# Distribution of Ethnic Group.

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l= ["groupA","groupB","groupC","groupD","groupE"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]


plt.pie(mylist,labels=l, autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
```
